```python
import numpy as np
import matplotlib.pyplot as plt

elementos = ['100X100', '200X200', '400X400', '800X800', '1600X1600']
dados = [54, 311, 2404, 20512, 206187]
colors = ['red', 'blue', 'green', 'purple', 'orange']


fig, ax = plt.subplots(figsize=(14,18))

plt.title('Tempo por execução sequencia')
plt.ylabel('Tempo em (ms)')

graf = ax.bar(elementos[0], dados[0], color=colors[0], label=dados[0], width=0.9)
graf = ax.bar(elementos[1], dados[1], color=colors[1], label=dados[1], width=0.9)
graf = ax.bar(elementos[2], dados[2], color=colors[2], label=dados[2], width=0.9)
graf = ax.bar(elementos[3], dados[3], color=colors[3], label=dados[3], width=0.9)
graf = ax.bar(elementos[4], dados[4], color=colors[4], label=dados[4], width=0.9)

plt.legend()

plt.show()
```
